```python
from rembg import remove
from PIL import Image, ImageEnhance
import matplotlib.pyplot as plt

# Store path of the image in the variable input_path
query_path = r"C:\Users\HP\Downloads\WhatsApp Image 2024-04-15 at 7.57.59 AM (1).jpeg"
reference_path=r"C:\Users\HP\Downloads\WhatsApp Image 2024-04-15 at 7.57.59 AM.jpeg"

# Processing the image

input_q_img = Image.open(query_path)
input_r_img = Image.open(reference_path)
def enhance(input_img):
    # Decrease contrast and brightness
    
    enhancer = ImageEnhance.Contrast(input_img)
    output_img_with_less_contrast = enhancer.enhance(1.9)  # Decrease contrast by 20%
    
    enhancer = ImageEnhance.Brightness(output_img_with_less_contrast)
    output_img_final = enhancer.enhance(1)  # Decrease brightness by 20%
    # Removing the background from the given Image
    output_img = remove(output_img_final)
    
    # Converting transparent areas to white
    output_img_with_white_bg = Image.new("RGB", output_img.size, (255, 255, 255))
    output_img_with_white_bg.paste(output_img, (0, 0), output_img)

    return output_img_with_white_bg
    
query_img=enhance(input_q_img)
reference_img=enhance(input_r_img)

# plt.imshow(output)
# plt.axis('off')  # Hide axis
# plt.show()
```
